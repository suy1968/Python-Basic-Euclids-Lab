# Seaborn

Seaborn is an open source, Python library providing high level API for visualizing the data using Python programming language. It is built on top of Matplotlib.

Important Features of Seaborn
1. Built in features for styling matplotlib.
2. Easy to visualize univariate and multivariate analysis.
3. Easy to work with Pandas dataframe and Series.
4. Comes with bult-in datasets.

**Importing all the required library**

Lets import Pandas and Numpy for dataframe and data manipulation. Matplotlib for ploting. Seaborn for highlevel API for data visualization

```python
#Import Pandasfor datamanipulations and analysis
#Import numpy for numeric calculation
#Import Matplotlib.pyplot and seaborn for datavisualization

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Check built-in datasets in seaborn
sns.get_dataset_names()
```

```python
#Load iris dataset from seaborn and check the head
df = sns.load_dataset('iris')
df.head()
```

### Distribution Plots

#### Univariate analysis

**Distplot**

Histograms represent the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

sns.distplot(
    a,
    bins=None,
    hist=True,
    kde=True,
)

* bins: The intervals into which the entire range of the data is split into is called as the bin. 
* kde:  (Kernal density estimation) It will estimate  the probability density function for the given data.

```python
#Distribution plot for sepal_width
sns.distplot(df['sepal_width'])
```

```python
#Distribution plot for sepal_width, set kde parameter false
sns.distplot(df['sepal_width'],kde=False)
```

#### Bivariate Analysis

```python
#Plot jointplot for 'petal_length' and 'petal_width'
sns.jointplot(x = 'petal_length',y = 'petal_width',data = df)
```

```python
#Plot jointplot for 'petal_length' and 'petal_width', use kind = 'kde'
#kind : { "scatter" | "reg" | "resid" | "kde" | "hex" }, optional. Try to use all of them
sns.jointplot("petal_length", "petal_width", data=df,kind="kde", space=0)
```

**lmplot()**

```python
#Plot lmplot for 'petal_length' and 'petal_width' using hue as species. Play with fit_reg
sns.lmplot(x='petal_length', y='petal_width', hue='species', data=df, fit_reg=False)
```

### (n,2) combination of Bivariate Analysis

We can observe the variations in each plot. The plots are in matrix format where the row name represents x axis and column name represents the y axis.

```python
#Plot pairplot matrix with hue as species, diag_kind 'kde',kind = "scatter" and husl palette
sns.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
```

### Ploting Categorical Data

When one or both the variables are categorical, we use plots like ``striplot(), swarmplot()``, etc,. Seaborn provides interface to do so.

When datapoints are less and well distributed use `stripplot()`. When its concentrated its better to use `swarmplot()`.

Swamplot is also the optional for Jitter in stripplot.

**Stripplot**

```python
#Jitter (only along the categorical axis) to apply. This can be useful when you have many points and they overlap, so that it is easier to see the distribution.
#Plot stripplot for species categories and analyse their 'petal_length'. What if Jitter is false or true?
sns.stripplot(x = "species", y = "petal_length", data = df,jitter=False)
```

**SwarmPlot**

Tt is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
#Plot searmplot for species categories and analyse their 'petal_length'.
sns.swarmplot(x = "species", y = "petal_length", data = df)
```

### Ploting Categorical Data Distribution



**Box Plot**

Boxplot is a convenient way to visualize the distribution of data through their quartiles.

Box plots usually have vertical lines extending from the boxes which are termed as whiskers. These whiskers indicate variability outside the upper and lower quartiles, hence Box Plots are also termed as box-and-whisker plot and box-and-whisker diagram. Any Outliers in the data are plotted as individual points.

```python
#Plot boxplot for species categories and analyse their 'petal_length'.
sns.boxplot(x = "species", y = "petal_length", data = df)
```

**Violin plot**

Violin Plots are a combination of the box plot with the kernel density estimates. So, these plots are easier to analyze and understand the distribution of the data.

```python
#Plot violinplot for species categories and analyse their 'petal_length'.
sns.violinplot(x = "species", y = "petal_length", data = df)
```

### Statistical Estimation

Here we deal with central tendency estimation, we need a specific way to summarize the distribution. Mean and median are the very often used techniques to estimate the central tendency of the distribution.

**Bar Plot**

The barplot() shows the relation between a categorical variable and a continuous variable. The data is represented in rectangular bars where the length the bar represents the proportion of the data in that category.

Bar plot represents the estimate of central tendency. 

```python
#Plot barplot for species categories and analyse their 'petal_length'.
sns.barplot(x = "species", y = "petal_length", data = df)
```

```python
#Plot barplot for species categories and analyse their 'petal_length' use standard deviation estimator 
import numpy as np
sns.barplot(x = "species", y = "petal_length", data = df, estimator=np.std)
```

**Count Plot**

Instead of central tendency, here we deal with the total number of observations.

```python
#Use countplot for species for total number of observations
sns.countplot(x="species",data=df)
```

### Facet Grid

FacetGrid class helps in visualizing distribution of one variable as well as the relationship between multiple variables separately within subsets of your dataset using multiple panels.

A FacetGrid can be drawn with up to three dimensions − row, col, and hue.

```python
#maping two plots barplot for 'petal_length' and 'petal_width' w.r.t 'species'
grid = sns.FacetGrid(data=df,col="species", )
grid.map(sns.barplot, "petal_length",color= 'green')
grid.map(sns.barplot, "petal_width",color='red')
```

## Matrix Plot

**Heatmap**

Heatmap is defined as a graphical representation of data using colors to visualize the value of the matrix. 

**Syntax: seaborn.heatmap(data, *, vmin=None, vmax=None, cmap=None, center=None, annot_kws=None, linewidths=0, linecolor=’white’, cbar=True, **kwargs)** 

Parameters:

* data: 2D dataset that can be coerced into an ndarray.
* vmin, vmax: Values to anchor the colormap, otherwise they are inferred from the data and other keyword arguments.
* cmap: The mapping from data values to color space.
* center: The value at which to center the colormap when plotting divergent data.
* annot: If True, write the data value in each cell.
* fmt: String formatting code to use when adding annotations.
* linewidths: Width of the lines that will divide each cell.
* linecolor: Color of the lines that will divide each cell.
* cbar: Whether to draw a colorbar.


```python
#Load flights dataset from seaborn
flights = sns.load_dataset('flights')
flights.head(3)
```

```python
flights.year.unique()
```

```python
flights.info()
```

```python
#Create a spreadsheet-style pivot table as a DataFrame using pivot_table(). 
#Set values as 'passengers', columns as 'year' and index as 'month'
flight_matrix = flights.pivot_table(values='passengers',columns='year',index='month')
```

```python
#Plot heatmap. If heatmap is too small then use figsize to set the figure size for plot
fig, ax = plt.subplots(figsize=(15,15))  
sns.heatmap(flight_matrix, ax=ax,cmap="YlGnBu",linewidths=0.5, annot=True, fmt= '.2f')
```
